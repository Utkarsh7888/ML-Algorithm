```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head(2)
```

```python
df.shape
```

```python
X = df.iloc[:,2:4].values
```

```python
y = df.iloc[:,-1].values
```

```python
y
```

```python
X.shape
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
np.sqrt(X_train.shape[0])
```

```python
k=5
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=2, metric='minkowski', p=2 ) 
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
error_train=[]
error_test=[]

for i in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    x=confusion_matrix(y_train,knn.predict(X_train))
    y=confusion_matrix(y_test,knn.predict(X_test))
    error_train.append((x[0][1]+x[1][0])/x.sum())
    error_test.append((y[0][1]+y[1][0])/y.sum())
```

```python
plt.plot(range(1,26),error_train,label='training error rate')
plt.plot(range(1,26),error_test,label='validation error rate')
plt.xlabel('K label')
plt.ylabel('Error')
plt.legend()
```

```python
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
def predict_output():
    age=int(input("Enter the age"))
    salary=int(input("Enter the salary"))
    
    X_new=np.array([[age],[salary]]).reshape(1,2)
    X_new=scaler.transform(X_new)
    
    if knn.predict(X_new)[0]==0:
        return"No you Will not purchase"
    else:
        return"Yes Will purchase"
```
